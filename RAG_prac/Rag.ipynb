```python
import wikipediaapi
import os
```

```python

wiki = wikipediaapi.Wikipedia(
    user_agent = "WW2 RAG",
    language = "en" 
)

page = wiki.page("World War II")

if(page):
    print("OK it exists!!")
else:
    print("No such page exists!!")    

```

```python
data = page.text
print(len(data))
print()
```

```python
metadata = {
    "title" : page.title,
    "source" : "Wiki",
    "language" : "en"
    
}
```

```python
with open("data.txt" , "w" , encoding="utf-8") as f:
    f.write(data)

parent_file = r"c:\Users\SIDDHAARTH\Desktop\RAG_prac"
file = r"data.txt"

checker = os.path.join(parent_file , file)
if os.path.isfile(checker):
    print("File is in the right directory!!")
else:
    print("Error")    
```

```python
chunck_size = 100
chunks = []
for i in range(0 , len(data) , chunck_size):
    chunks.append(data[i:i+chunck_size])

    
```

```python
chunk_info = []
for index , text in enumerate(chunks):
    chunk_info.append({
        "chunk_id" : index,
        "data" : text,
        "source" : metadata["title"]
    })
```

```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer("BAAI/bge-base-en-v1.5")
embeding = model.encode(chunks)
```
